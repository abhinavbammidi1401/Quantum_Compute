```python
import numpy as np
from qiskit import QuantumCircuit
from qiskit_aer import Aer
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector, plot_histogram

# Step 1: Tensor Product of Two Vectors
vector_0 = np.array([5, 4])  # |0⟩
vector_1 = np.array([9, 3])  # |1⟩

tensor_product = np.kron(vector_0, vector_1)
print("Tensor product of |0⟩ and |1⟩:", tensor_product)

# Step 2: Create an Operator and Apply on State Vector
qc = QuantumCircuit(2)  # 2 qubits
qc.h(0)  # Hadamard on qubit 0
qc.x(1)  # Pauli-X on qubit 1

# Display the circuit
qc.draw('mpl')

# Initialize state |00⟩
state = Statevector.from_label('00')
new_state = state.evolve(qc)

# Visualize the new state on Bloch Sphere
plot_bloch_multivector(new_state)

# Step 3: Partial Measurement
qc.measure_all()  # Measure both qubits

# Run the experiment
backend = Aer.get_backend('qasm_simulator')
job = backend.run(qc, shots=10000)
result = job.result()

# Get the measurement result
counts = result.get_counts()

# Plot the histogram of results
plot_histogram(counts)
```
